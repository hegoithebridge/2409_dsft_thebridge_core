# Extrae los dominios de los correos con una expresi√≥n regular

```python
import re

emails = '''
miguelnievas@thebridgeschool.tech
miguel.nievas@uc3m.es
miguel-12-1993@mi-correo.net
miguelniev@s@thebridgeschool.es
'''
```

Es decir.. 
* @thebridgeschool.tech
* @uc3m.es
* @mi-correo.net
* @thebridgeschool.es

```python
pattern = re.compile("@.+")

matches = pattern.finditer(emails)

for match in matches:
    print(match)
```

```python
pattern = re.compile("@[\w-]+\.\w+")

matches = pattern.finditer(emails)

for match in matches:
    print(match)
```

```python
pattern = re.compile("@[a-zA-Z0-9-]+\.[a-zA-Z0-9-.]+")

matches = pattern.finditer(emails)

dominios = []
for match in matches:
    # print(match.group())
    dominios.append(match.group())
dominios
```

```python
matches = pattern.finditer(emails)
dominios = [match.group() for match in matches]
dominios
```
