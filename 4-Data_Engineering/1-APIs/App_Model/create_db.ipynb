```python
import sqlite3
import pickle
```

```python
connection = sqlite3.connect('ejercicio/data/advertising.db')
```

```python
import pandas as pd

df = pd.read_csv('data/Advertising.csv', index_col=0)
df.head()
```

```python
df.to_sql(name="campañas", con=connection, index=False)
```

```python
cursor = connection.cursor()
select_books = "SELECT * FROM campañas"
result = cursor.execute(select_books).fetchall()
print(result)
```

```python
df = pd.DataFrame(data= result, columns= ['tv', 'radio', 'newspaper', 'sales'])
```

```python
df
```

```python
model = pickle.load(open('data/advertising_model', 'rb'))
```

```python
model.predict([[1, 2, 3]])
```

```python
model.fit(df[['tv', 'radio', 'newspaper']], df['sales'])
```

```python
for i in cursor.description:
    print(i[0])
```

```python
from sklearn.model_selection import cross_val_score
```

```python
abs(cross_val_score(model, df.iloc[:, :-1].dropna(), df.iloc[:, -1].dropna(), scoring= 'neg_mean_absolute_error', cv = 5).mean())
```
